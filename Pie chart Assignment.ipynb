```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv('dataset.csv')
df
```

```python
country=df['country'].values
gold_medal=df['gold_medal'].values
color=['red','green','blue','gold','purple']
arr=(0,0.2,0,0,0)
plt.pie(gold_medal,labels=country,colors=color,autopct='%.1f%%',explode=arr,shadow=True)
plt.legend(country,loc="lower left")
plt.axis('equal')
plt.show()


```
