```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('computer_parts.csv')
df
```

```python
product=df['Product Name;'].values
numberOfBugs=df['Number Of Bugs'].values
plt.pie(numberOfBugs,labels=product,autopct='%.1f%%',shadow=True,radius=15)
plt.legend(product,loc='upper center')
plt.title('Data')
plt.show()
```
